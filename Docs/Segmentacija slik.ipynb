## Pristopi



###Pragovna segmentacija 

Pragovna segmentacija je preprosta oblika binarne segmentacije - sliko razdelimo v dve regiji.

1. izberemo prag $v_t$
2. če $I[r,c] < v_t$ potem segment 1
3. če $I[r,c] \ge v_t$ potem segment 2

Prednosti:
* hitra in preprosta metoda
Slabosti:
* izbrati moramo primeren prag
* sliko lahko segmentiramo le na dva dela
* osnovni podatki pogosto niso deljivi v dva razreda s pragom

Kako izberemo prag?

#### Metoda Otsu

https://en.wikipedia.org/wiki/Otsu's_method

Izbran prag naj minimizira varianco vrednosti v obeh segmentih. Takšen prag poiščemo z izčrpnim iskanjem:

1. uredimo vse vrednosti po vrsti
2. vsako izmed unikatnih vrednosti preizkusimo kot prag
3. kot prag izberemo vrednost z najmanjšo varianco

Metodo lahko aproksimiramo z uporabo histograma.

```python
# slika z WikiMedie Angkor_Sunrise_Silhouette.JPG
slika = pyplot.imread('./sunrise.jpg').mean(2)
pyplot.imshow(slika, cmap='gray')
pyplot.axis('off');
```

```python
v1 = np.random.randn(100)*0.5+1
v2 = np.random.randn(100)*0.2

podatki = slika.ravel()
hist_val,hist_ind = np.histogram(podatki,100)
hist_width = hist_ind[1]-hist_ind[0]
hist_ind = hist_ind[:-1]
bar(hist_ind, hist_val, hist_width)
```

```python
seg_var_total = np.zeros(hist_ind.shape)

values = hist_ind+hist_width/2.
counts = hist_val

seg_var_total[0] = (values**2*counts).sum()/counts.sum()
seg_var_total[0] -= ((values*counts).sum()/counts.sum())**2
for i in range(1, values.shape[0]):
    seg1_mean = (values[:i]*counts[:i]).sum()/counts[:i].sum()
    seg1_sq_mean = (values[:i]**2*counts[:i]).sum()/counts[:i].sum()
    
    seg2_mean = (values[i:]*counts[i:]).sum()/counts[i:].sum()
    seg2_sq_mean = (values[i:]**2*counts[i:]).sum()/counts[i:].sum()
    
    seg1_var = seg1_sq_mean-seg1_mean**2
    seg2_var = seg2_sq_mean-seg2_mean**2
    seg_var_total[i] = seg1_var+seg2_var

otsu_prag = values[np.argmin(seg_var_total)]

pyplot.subplot(121)
pyplot.plot(values, seg_var_total)
pyplot.axvline(otsu_prag)

pyplot.subplot(122)
pyplot.bar(hist_ind, hist_val, hist_width)
pyplot.axvline(otsu_prag)

pyplot.figure()
pyplot.imshow(slika>otsu_prag)
pyplot.axis('off')

print('izbran prag {:.2f}'.format(otsu_prag))
```

###KMeans segmentacija

https://en.wikipedia.org/wiki/K-means_clustering

Sliko segmentiramo z metodo K-povprečij.

Prednosti:
* algoritem je preprost za implementacijo
* izbrati moramo samo število segmentov

Slabosti:
* preprosta implementacije niso učinkovite
* regije niso nujno povezane (odvisno od značilk)

```python
# slika iz zbirke Berkeley Segmentation Dataset
slika = pyplot.imread('101087.jpg')/255.
pyplot.imshow(slika)
pyplot.axis('off')
```

```python
K = 10
podatki = slika.reshape(-1, 3)
povprecja_init = np.random.permutation(podatki.shape[0])[:K]
povprecja = podatki[povprecja_init,:]
segment = np.zeros(podatki.shape[0])
razdalja = np.zeros((podatki.shape[0],K))
for itt in range(0,100):
    for k in range(0,K):
        razdalja[:,k] = ((podatki - povprecja[k,:].reshape(1,3))**2).sum(1)**0.5
    segment = np.argmin(razdalja, 1)

    nova_povprecja = povprecja.copy()
    for k in range(0,K):
        if (segment==k).max() == False:
            continue
        nova_povprecja[k,:] = podatki[segment==k,:].mean(0)
    
    # lahko izpisujemo spremembo povprecji in spremljamo konvergenco
    # pametnejsi algoritem bi zakljucil ko bi bila razlika dovolj majhna
    # print('sprememba povprecji {:f}'.format(((povprecja-nova_povprecja)**2).mean()**0.5))
    povprecja = nova_povprecja
```

```python
pyplot.subplot(121)
# sliko segment preoblikujemo v obliko slike za prikaz
# uporabimo samo prvi dve dimenziji
pyplot.imshow(segment.reshape(slika.shape[:2]), cmap='gray')
pyplot.axis('off')
pyplot.subplot(122)
# povprecja indeksiramo z indeksi segmentov
# za vsako vrednost v segment dobimo pripadajoco vrednost iz 
# povprecja, nato preoblikujemo v barvno sliko
pyplot.imshow(povprecja[segment,:].reshape(slika.shape))
pyplot.axis('off')
```

###Segmentacija z grafi

Piksle v sliki lahko predstavimo z grafom, kjer piksli predstavljajo vozlišča. Uteži povezav v grafu predstavljajo razliko med povezanimi vozlišči - piksli iz enakih regij imajo nizko vrednost, piksli iz različnih pa visoko. 

http://cs.brown.edu/~pff/segment/



```python
# slika iz zbirke Berkeley Segmentation Dataset
slika = pyplot.imread('./101087.jpg')
# algoritem je pocasen, zato pomanjsamo sliko
slika = scipy.misc.imresize(slika,0.25)/255.

pyplot.imshow(slika)
pyplot.axis('off')

H,W = slika.shape[:2]
```

Najprej generiramo vse povezave v grafu - ne rabimo povezati vseh pikslov, samo tiste v neki smiselni soseščini. Med inicializacijo vsakemu vozlišču določimo tudi svoj segment.

Tukaj bomo uporabili 4-soseščino. Vsak piksel bo povezan z sosedom zgoraj in spodaj ter levo in desno.



```python

vrtica_ind, stolpec_ind = np.meshgrid(np.arange(0,slika.shape[0]),
                                      np.arange(0,slika.shape[1]),
                                      indexing='ij')

N = H*W
indeksi_vozlisc = np.arange(0,N).reshape(H,W)

razdalja_dol = ((slika[:-1,:,:]-slika[1:,:,:])**2).sum(2)**0.5
razdalja_desno = ((slika[:,:-1,:]-slika[:,1:,:])**2).sum(2)**0.5

# vsako vozlisce bo imelo dve povezavi - z sosedom desno in sosedom spodaj
# izpustiti moramo samo zadnjo vrstico pri povezavah dol
# in zadnji stolpcev pri povezavah desno
povezave = np.zeros((N-W+N-H,3)) 

povezave[:N-W,0] = indeksi_vozlisc[:-1,:].reshape(-1)
povezave[:N-W,1] = indeksi_vozlisc[1:,:].reshape(-1)
povezave[:N-W,2] = razdalja_dol.reshape(-1)

povezave[N-W:,0] = indeksi_vozlisc[:,:-1].reshape(-1)
povezave[N-W:,1] = indeksi_vozlisc[:,1:].reshape(-1)
povezave[N-W:,2] = razdalja_desno.reshape(-1)

# zacetna segmentacija, vsako vozlisce v svojem segmentu
segmentacija = np.arange(0,N) 
seg_povezav = np.zeros((N-W+N-H,2))
seg_povezav[:,0] = segmentacija[np.int32(povezave[:,0])]
seg_povezav[:,1] = segmentacija[np.int32(povezave[:,1])]
```

Povezave sortiramo. Nato za vsako povezavo od najmanjše do največje pogledamo, ali bomo združili segmente, v katerih sta njene vozlišča.

```python
sort_ind = np.argsort(povezave[:,2])# sortiramo povezave po utezi narascajoce
povezave = povezave[sort_ind,:]
```


Imejmo povezavo med vozliščema $v_i$ in $v_j$, utež povezave pa je $w(v_i,v_j)$. Če vozlišči nista v istem segmentu, potem segmenta združimo kadar j:
$$
w(v_i,v_j) \le max(int(S_i), int(S_j))
$$

$S_i$ in $S_j$ sta grafa - segmenta, ki ju želimo združiti. $Int$ pa je maksimalna utež povezav v minimalnem vpetem drevesu teh grafov (**pozor, to ni maksimalna utež povezav v tem grafu**). 


```python
# tukaj poiscemo maksimalno povezavo v vpetem drevesu
# v bistvu malce goljufamo - ker imamo povezave urejene po utezem
# poiscemo utez zadnje povezave, ki jo moramo dodati
# da povezemo vsa vozlisca
def int_fun(povezave):
    vozlisca = np.unique(povezave[:,:2].ravel())
    for n in range(0,povezave.shape[0]):
        m = vozlisca==povezave[n,0]
        m += vozlisca==povezave[n,1]
        vozlisca[m] = -1
        if vozlisca.max() == -1:
            Int_c_i = povezave[n,2]
            return Int_c_i
    
```

Ker je ta vrednost nestabilna kadar je število vozlišč majhno, dodamo še konstanto.
$$
w(v_i,v_j) \le max(int(S_i)+ {k \over |S_i| }, int(S_j)+ {k \over |S_j| })
$$

Spremenljivka $k$ je parameter metode s katerim nadzorujemo število tvorjenih segmentov. Delimo jo z številom vozlišč v grafu - segmentu. Kadar je število vozlišč v segmentu veliko, $k$ nima bistvenega vpliva na rezultat primerjave.

```python
k = 10.
```

```python
for q in range(0, povezave.shape[0]):
    v_i, v_j, w = povezave[q]
    s_i = segmentacija[v_i]
    s_j = segmentacija[v_j]
    # vozlisci sta ze v istem segmentu
    if s_i == s_j:
        continue
    
    c_i = segmentacija == s_i
    c_j = segmentacija == s_j
    
    c_i_seg = (seg_povezav == s_i).min(1)
    c_j_seg = (seg_povezav == s_j).min(1)
    
    Int_c_i = k/c_i.sum()
    if c_j_seg.sum() > 0:
        Int_c_i += int_fun(povezave[c_i_seg,:])
        
    Int_c_j = k/c_j.sum()
    if c_j_seg.sum() > 0:
        Int_c_i += int_fun(povezave[c_j_seg,:])
   
    MInt = np.min([Int_c_i,Int_c_j])

    if w <= MInt:
        seg_povezav[c_j_seg] = s_i
        segmentacija[c_j] = s_i
```

Indeksi segmentov so trenutno pomešani, kako so bili pač združevani v metodi. Tu jih še enkrat indeksiramo, lepo po vrsti od 0 naprej.

```python
sn = 0
for s in np.unique(segmentacija):
    segmentacija[segmentacija==s] = sn
    sn += 1
```

Segmente lahko preoblikujemo v sliko in jih pogledamo. Poglejmo tudi število generiranih segmentov.

```python
segmentacija = segmentacija.reshape(H,W)
pyplot.imshow(segmentacija)
pyplot.axis('off')

np.unique(segmentacija).shape
```

Pripravimo še sliko, kjer vsak segment pobarvamo z njegovo povprečno barvo.

```python
slika_seg = np.zeros(slika.shape)
for k in np.unique(segmentacija):
    slika_seg[segmentacija==k,:] = slika[segmentacija==k,:].mean(0, keepdims=True)
    
imshow(slika_seg)
pyplot.axis('off')
```

### Slic segmentacija
#### Simple Linear Iterative Clustering 

Metoda Slic je optimizirana različica K-povprečij:
http://ivrl.epfl.ch/research/superpixels





```python
slika = pyplot.imread('101087.jpg')/255.
```

KMeans lahko izboljšamo, če namesto samo barvnih vrednosti dodamo tudi lokacije pikslov. Tako so dobljeni segmenti bližje skupaj tudi v prostoru slike (ne samo barvnem prostoru). 

Prvi parameter algoritma je obtežitev prostorskih koordinat v primerjavi z barvami. Odvisno on naših želja optimizacije bomo želeli večjo pomembnost  dati prostorskim koordinatam ali pa barvnim vrednostim. 

```python
H,W = slika.shape[:2]
row_ind = np.arange(0,H)
col_ind = np.arange(0,W)
R,C = np.meshgrid(row_ind,col_ind,indexing='ij')
podatki = np.zeros((H*W,5))
podatki[:,0] = R.reshape(-1)/H # normaliziramo koordinatni prostor
podatki[:,1] = C.reshape(-1)/W # normaliziramo koordinatni prostor
podatki[:,2:] = slika.reshape(-1,3)
```

Pomembna je inicializacija povprečij segmentov. Namesto naključne izbire je te bolj smiselno enakomerno porazdeliti po prostoru v mreži. Drug parameter je dimenzija mreže s katero inicializiramo metodo.

Kot zadnjo izboljšavo pa lahko upotevamo še lokalnost posameznih segmentov. Piksli, ki so v nekem koraku daleč od segmenta, temu segmentu ne bodo dodani. Zato lahko pri ponovnem računanju upoštevamo samo
tiste piksle, ki so blizu trenutnemu segmentu.



Metoda daje dobre rezultate: na tej vpliva dodatek koordinat in inicializacija.
Metoda je hitra: za hitrost moramo optimizirati računanje povprečij in pripadnost segmentov na smiselne okolice.
